#### Q5. Show the car details as per buyers choice such as company, body-style, horsepower, average-mileage, price.


```python
while True:
    c= int(input("Enter 1 for company, 2 for body-style, 3 for horsepower, average mileage, 4 for price, 5 for exit "))
    if c==1:
        company= input("Enter company name ")
        print(df[df['company']==company])
    if c==2:
        bs= input("Enter body-style ")
        print(df[df['body-style']==bs])
    if c==3:
        hp= int(input("Enter horsepower "))
        print(df[df['horsepower']>=hp])
    if c==4:
        am= int(input("Enter average mileage "))
        print(df[df['average-mileage']>=am])
    else:
        break
```
